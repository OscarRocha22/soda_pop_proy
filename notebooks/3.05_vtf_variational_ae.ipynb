# <font color= #f6c049> **Modeling** </font>

The proposed architecture of the Variational Autoencoder (VAE) is less computationally expensive than previous iterations that feature a Perceptual Loss Function. For this reason, the image size will be 256x256 pixels. This results in the following dimensionality reduction:

$$
64,536_{input} \rightarrow 16,384_{h_1} \rightarrow 4096_{h_2} \rightarrow 1024_z \leftarrow 4096_{h_2} \leftarrow 16,384_{h_1} \leftarrow 64,536_{output}
$$

The hyperparameters of the Convolutional Perceptual Autoencoder will be:

- Optimizer: Adam
- Loss: Perceptual Loss
- Epochs: 5
