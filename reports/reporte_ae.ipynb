## <font color= #f6c049> **About The Data** </font>

This project aims to create an autoencoder neural network capable of generating images of soda cans. The images were scraped from Google Images using custom scripts to both obtain the images and download them. After that, we discarded images that were not going to be useful (such as images with busy backgrounds or others that just didn't show soda cans) by hand, ending up with approximately **1600** images for training, validation and testing.

We loaded the images so that all were **64x64** pixels in size. This shrunk some images and expanded others, and it also changed the aspect ratio of some wider images to be squared, and the opposite for skinny images. **80%** of the images were reserved for training, **10%** for validation and **10%** for testing. 


## <font color= #f6c049> **Implementations** </font>

In this project we present two different implementations of autoencoders for image generation. 

#### <font color= #f6c049> **Convolutional Autoencoder With A Perceptual Loss Function using VGG-19** </font>

VGG-19 is a convolutional neural network designed in 2015 by Karen Simonyan & Andrew Zisserman for image recognition. This neural network has an understanding of deep patterns of images that it got during training. The last convolutional layer of this model can be used as a loss function for our own NN. The weights of the VGG model were frozen so that they wouldn't be changed during our own training.

The dimensionality reduction performed by the neural network goes as follows:
$$
4096_{input}  \rightarrow 1024_{h_1} \rightarrow 256_{h_2} \rightarrow 64_z \leftarrow 256_{h_2} \leftarrow 1024_{h_1} \leftarrow 4096_{output}
$$

We used the Adam optimizer for training, our custom Perceptual Loss function obtained from VGG19 and trained for 5 epochs.

**RESULTS**

![CAE Results](../reports/figures/cae_vgg_results.png "CAE Results")

With only 5 epochs and less than 100,000 total parameters, this implementation can reconstruct the images in the testing dataset with very high accuracy. The new images are slightly more blurry than the originals but small details like the soda name and tag-lines are still legible, even though the images are very small.

#### <font color= #f6c049> **Variational Autoencoder** </font>

Variational Autoencoders see the latent space as smooth and probabilistic. Unlike our last approach, that works best for image reconstruction (remaking something that is already there) this approach will be better suited to try and generate brand new images that aren't in the test dataset.

The dimensionality reduction performed by the neural network goes as follows:
$$
4096_{input}  \rightarrow 1024_{h_1} \rightarrow 256_{h_2} \rightarrow 64_z \leftarrow 256_{h_2} \leftarrow 1024_{h_1} \leftarrow 4096_{output}
$$

This autoencoder uses KL-divergence and binary crossentropy as its loss functions. We used the Adam optimizer for training and trained for 30 epochs.

**RESULTS**

![](../reports/figures/vae_results_1.png) ![](../reports/figures/vae_results_2.png)

The images generated by the decoder are very blurry, with only the general shape of the soda can being recognizeable. The colors of the cans are also washed and non-specific, a consequence of the original dataset having many different colored cans (green, red, blue, etc) which appear to have merged together in the generations.

## <font color= #f6c049> **Challenges We Faced** </font>

Our first small obstacle was obtaining the data itself. We couldn't find a pre-made dataset of soda cans so we had to make our own, which taught us a lot about web scraping. 

We also had an issue with computer resources. We originally intended to use **256x256** images for our model but both GPUs available to us didn't have enough VRAM to handle that amount of images at that size, which led us to using 64x64 images instead. This might have been a blessing in disguise, as most images were originally a lot closer to 64x64 than 256x256 in size, so in the end, the images were barely changed from their original form.

## <font color= #f6c049> **Potential Applications** </font>

Our findings could be useful in different soda and non-soda related fields. A [bottler](https://en.wikipedia.org/wiki/Bottling_company) could use our CAE to quickly check for imperfections, defects or differences in the cans they produce, as a form of **quality control**. With the appropiate training, this could also be used for quality control of bottles as well, not just cans.

Our image generation approach with the VAE is still primitive but relatively light on resources (it can generate many images in less than a second). With more fine-tuning, this VAE could help for brainstorming of custom cans for special promotions, or generating any other image in general with the proper training dataset.
